```python
from fastai.vision import *
```

```python
PATH = Path("/home/vhl_ai/Data/Alex")
```

```python
chat = PATH / "chat.csv"
```

```python
df = pd.read_csv(chat, header = None)
```

```python
df.columns = ['date']
```

```python
date = df.date
```

```python
def ordered_set(sequence):
    seen = set()
    return [i for i in sequence if not (i in seen or seen.add(i))]
```

```python
set_date = ordered_set(date)
```

```python
freq = df.groupby(['date'], sort=False).size()
```

```python
freq = [i for i in freq]
```

```python
fin = pd.DataFrame({"date": set_date, "freq": freq})
```

```python
index = []
for i,o in enumerate(fin['date']):
    if '/' not in str(o):
        index.append(i)
fin = fin.drop(fin.index[[index]])
```

```python
ax = fin.plot.bar(x='date', y='freq', figsize=(80,20), color='green')
ax.bar(x='date',width=1, height=1)
ax.tick_params(axis='x', rotation=45)
ax.get_legend().remove()
ax.tick_params(axis='both',labelsize='large')
ax.xaxis.set_label_text('')
temp = ax.xaxis.get_ticklabels()
temp = list(set(temp) - set(temp[::7]))
for label in temp:
    label.set_visible(False)
```
